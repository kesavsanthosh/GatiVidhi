```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import os
import time

# URL of the website
website_url = "https://indiankanoon.org"  # Replace this with the actual website URL

# Keywords to search
keywords = ["bajaj", "medical", "third party claim"]  # Replace these with your desired keywords

# Combine the keywords into a single search query
search_query = " ".join(keywords)

# Create a new folder to store the downloaded PDFs
download_directory = r"C:/Users/Venkata Vignesh/Documents/hackrx/pdfs"  # Replace this with the desired folder path
os.makedirs(download_directory, exist_ok=True)

# Function to download a PDF file
def download_pdf(url, filename):
    driver.get(url)
    wait = WebDriverWait(driver, 10)
    try:
        # Wait for the "Get this document in PDF" button to be clickable
        get_pdf_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//input[@value='Get this document in PDF']")))
        get_pdf_button.click()
        time.sleep(5)  # Add a delay to allow time for the download to complete
    except Exception as e:
        print(f"Error downloading PDF: {e}")
        return

# Start a Selenium session with Chrome
driver = webdriver.Chrome()
driver.get(website_url)

# Find the search input element and enter the combined search query
search_input = driver.find_element("id", "search-box")  # Replace "search-box" with the actual ID of the search input element
search_input.clear()
search_input.send_keys(search_query)
search_input.send_keys(Keys.RETURN)

# Wait for the search results page to load
time.sleep(5)

# Get the page source and parse it with BeautifulSoup
page_source = driver.page_source
soup = BeautifulSoup(page_source, "html.parser")

# Find all links on the page
links = soup.find_all("a")

# Loop through the links to find and download PDF files that have "/doc/" in the link
for link in links:
    href = link.get("href")
    if "/doc/" in href:
        pdf_url = f"https://www.indiankanoon.org{href}"
        pdf_filename = os.path.join(download_directory, os.path.basename(pdf_url) + ".pdf")
        download_pdf(pdf_url, pdf_filename)
        print(f"Downloaded: {pdf_filename}")

# Print a message indicating the completion of the keyword search
print(f"Completed keyword search for: {search_query}")

# Close the browser
driver.quit()
```
